```python
import pandas as pd
```

```python
df_dim = pd.read_csv('dim_datasets.csv')
df_pandas = pd.read_csv('fact_metrics_pandas.csv')
df_pyspark = pd.read_csv('fact_metrics_pyspark.csv')
df_polars = pd.read_csv('fact_metrics_polars.csv')
df_pyarrow = pd.read_csv('fact_metrics_pyarrow.csv')
```

```python
display(df_dim[['dataset_name', 'number_of_rows']].sort_values('number_of_rows', ascending=False))
```

```python
result = pd.concat([
    df_pandas.assign(
        dataset_id=df_pandas['dataset_id'].map({
            1: 'transactions_data_pandas',
            2: 'locations_pandas',
            3: 'titanic_pandas',
            4: 'reviews_pandas'
        }).fillna('Unknown')
    ),
    df_pyspark.assign(
        dataset_id=df_pyspark['dataset_id'].map({
            1: 'transactions_data_spark',
            2: 'locations_spark',
            3: 'titanic_spark',
            4: 'reviews_spark'
        }).fillna('Unknown')
    ),
    df_polars.assign(
        dataset_id=df_pyspark['dataset_id'].map({
            1: 'transactions_data_polars',
            2: 'locations_polars',
            3: 'titanic_polars',
            4: 'reviews_polars'
        }).fillna('Unknown')
    ),
    df_pyarrow.assign(
        dataset_id=df_pyspark['dataset_id'].map({
            1: 'transactions_data_pyarrow',
            2: 'locations_pyarrow',
            3: 'titanic_pyarrow',
            4: 'reviews_pyarrow'
        }).fillna('Unknown')
    ),
], ignore_index=True)

display(result)
```
